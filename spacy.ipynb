```
# morphology and syntax
for token in doc:
    # Get the token text, part-of-speech tag and dependency label
    token_text = token.text
    token_pos = token.pos_
    token_tag = token.tag_
    token_lemma = token.lemma_
    token_dep = token.dep_
    # This is for formatting only
    print(f"{token_text:<12}{token_pos:<10}{token_tag:<10}{token_lemma:<20}{token_dep:<20}")
```

```
# TODO modify the code above to print the description of each tag

print(f"{'TEXT':<15} {'LEMMA':<15} {'POS':<10} {'POS DESC':<30} {'TAG':<10} {'TAG DESC':<40} {'DEP':<15} {'DEP DESC':<30}")
print("-" * 150)

for token in doc:
    # Get the token information
    token_text = token.text
    token_pos = token.pos_
    token_tag = token.tag_
    token_lemma = token.lemma_
    token_dep = token.dep_

    # Get the descriptions
    pos_desc = spacy.explain(token_pos) if spacy.explain(token_pos) else ''
    tag_desc = spacy.explain(token_tag) if spacy.explain(token_tag) else ''
    dep_desc = spacy.explain(token_dep) if spacy.explain(token_dep) else ''

    # Print in a formatted table
    print(f"{token_text:<15} {token_lemma:<15} {token_pos:<10} {pos_desc:<30} {token_tag:<10} {tag_desc:<40} {dep_desc:<15} {dep_desc:<30}")
```

```
# TODO Iterate over the predicted entities
for ent in doc.ents:
    print(ent.text, ent.label_, spacy.explain(ent.label_))

```

+ TODO: print only the verbs, 3rd person singular present and the proper singular nouns

```
# TODO remove stopwords
filtered_tokens = []

for token in doc:
  if token.is_stop:
    continue
  filtered_tokens.append(token)

print(filtered_tokens)
```

```

# TODO print only verbs, 3rd person singular present and proper singular nouns

third_person_singular_present_verbs = []
nouns = []
verbs = []

for token in doc:
    # 3rd person singular present verbs
    if token.tag_ == "VBZ":
        third_person_singular_present_verbs.append(token)

    # Proper singular nouns
    if token.tag_ == "NNP":
        nouns.append(token)

    # All other verbs (excluding VBZ)
    if token.pos_ == "VERB" and token.tag_ != "VBZ":
        verbs.append(token)

# Print results
print("Proper singular nouns:", [token.text for token in nouns])
print("3rd person singular present verbs:", [token.text for token in third_person_singular_present_verbs])
print("Other verbs:", [token.text for token in verbs])
```
