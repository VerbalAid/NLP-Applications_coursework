The vectors are not the same because `sentence` was created without a tokenizer, so the final
full stop stays attached: `sentence[11]` is `"Washington."` not `"Washington"`. GloVe treats
them as different words, giving different vectors. Using `tokenized_sentence` fixes this by
correctly splitting off the punctuation, making both tokens identical.

````
# TODO: obtain new representation using the glove embeddings and compare them again.

# The original 'sentence' object had 'Washington.' as a token (sentence[11]),
# which is different from 'Washington' (sentence[0]).
# The 'tokenized_sentence' object, created with Flair's default tokenizer,
# correctly separates the punctuation, making both 'Washington' tokens identical.
# Specifically, tokenized_sentence[0] is 'Washington' and tokenized_sentence[13] is 'Washington'.

# Apply the embeddings to the tokenized_sentence
en_embedding.embed(tokenized_sentence)

# Compare the embeddings of the two 'Washington' tokens
import torch

if torch.equal(tokenized_sentence[0].get_embedding(), tokenized_sentence[13].get_embedding()):
  print("The vectors for 'Washington' (tokenized_sentence[0]) and 'Washington' (tokenized_sentence[13]) are now the same.")
else:
  print("The vectors are still not the same. Something is wrong with the tokenization.")
````

````
from segtok.segmenter import split_single

# 1. Open the text file
with open('/content/drive/My Drive/Colab Notebooks/2026-ILTAPP/2026-ILTAPP/resources/movie-review.txt', 'r') as f:
    docText = f.read()

# 2. Predict NER and Sentiment for the whole document
doc_sentence = Sentence(docText)
ner_tagger.predict(doc_sentence)
polarity_classifier.predict(doc_sentence)

print("=" * 60)
print("DOCUMENT-LEVEL ANNOTATIONS")
print("=" * 60)

doc_sentiment = [l for l in doc_sentence.labels if l.data_point == doc_sentence]
print(f"Sentiment : {doc_sentiment[0].value} ({doc_sentiment[0].score:.4f})")
print("Named Entities:")
for entity in doc_sentence.get_labels('ner'):
    print(f"  [{entity.data_point.text}] → {entity.value} ({entity.score:.4f})")

# 3. Segment document into sentences
sentences = [Sentence(s) for s in split_single(docText) if s.strip()]

for sent in sentences:
    ner_tagger.predict(sent)
    polarity_classifier.predict(sent)

# 4. Print sentence-level annotations
print("\n" + "=" * 60)
print("SENTENCE-LEVEL ANNOTATIONS")
print("=" * 60)

for sent in sentences:
    sent_sentiment = [l for l in sent.labels if l.value in ('POSITIVE', 'NEGATIVE')]
    entities = sent.get_labels('ner')

    print(f"\n  TEXT     : {sent.text}")
    if sent_sentiment:
        print(f"  Sentiment: {sent_sentiment[0].value} ({sent_sentiment[0].score:.4f})")
    if entities:
        ent_str = ", ".join([f"{e.data_point.text} ({e.value})" for e in entities])
        print(f"  Entities : {ent_str}")
    else:
        print(f"  Entities : none")
    print("  " + "-" * 56)
````

## Assignment 4 – Observations: Document-level vs. Sentence-level Annotations

**Sentiment:**
At document level, the whole review gets one label: NEGATIVE (1.0). At sentence level, most sentences are also `NEGATIVE`, but two are ` POSITIVE`   because in isolation they sound neutral or factual. Document-level is more reliable for overall tone; sentence-level picks up local variation but can be misled by context.

**Named Entity Recognition:**
Both levels find the same entities (Costner, Ashton Kutcher, Kutcher) all tagged as PER. The main difference is that document-level lists all occurrences together (including duplicates), while sentence-level only shows entities found in that specific sentence.
